```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step 1:** Load the dataset

```python
#Load dataset
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
# Set the seed for reproducibility
np.random.seed(42)
```

```python
# Take a random sample of 25 observations
sample = df.sample(n=25, random_state=42)
```

## **Question B**

**To find the 98th percentile of BMI for both the sample and the population, we can follow these steps:**

* Calculate the 98th percentile of BMI for the sample.

* Calculate the 98th percentile of BMI for the entire population.

* Create a chart to compare these percentiles.

### **Step 1:** Calculate the 98th percentile of BMI for the sample and the population

```python
import numpy as np

# Sample data
sample_98th_percentile_bmi = np.percentile(sample['BMI'], 98)

# Population data
population_98th_percentile_bmi = np.percentile(df['BMI'], 98)

# Print the results
print("Sample 98th percentile BMI:", sample_98th_percentile_bmi)
print("Population 98th percentile BMI:", population_98th_percentile_bmi)
```

```python
import matplotlib.pyplot as plt

# Data for plotting
categories = ['98th Percentile of BMI']  # Define the category for the x-axis
sample_values = [sample_98th_percentile_bmi]  # Sample 98th percentile BMI value
population_values = [population_98th_percentile_bmi]  # Population 98th percentile BMI value

x = np.arange(len(categories))  # Create an array of positions for the category on the x-axis
width = 0.35  # Define the width of the bars

# Plotting the comparison bar charts
fig, ax = plt.subplots()  # Create a figure and a set of subplots
bars1 = ax.bar(x - width/2, sample_values, width, label='Sample')  # Bar chart for sample value
bars2 = ax.bar(x + width/2, population_values, width, label='Population')  # Bar chart for population value

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('Statistics')  # Label for the x-axis
ax.set_ylabel('BMI Values')  # Label for the y-axis
ax.set_title('Comparison of 98th Percentile of BMI between Sample and Population')  # Title of the plot
ax.set_xticks(x)  # Set the positions of the x-ticks
ax.set_xticklabels(categories)  # Set the label for the x-tick
ax.legend()  # Add a legend to the plot

# Function to add value labels on the bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()  # Get the height of the bar
        ax.annotate('{}'.format(round(height, 2)),  # Format the label with the bar height, rounded to 2 decimal places
                    xy=(bar.get_x() + bar.get_width() / 2, height),  # Set the x and y position for the label
                    xytext=(0, 3),  # Vertical offset of 3 points
                    textcoords="offset points",  # Specify the text coordinates
                    ha='center', va='bottom')  # Center horizontally and align text at the bottom vertically

# Add value labels to both sets of bars
add_value_labels(bars1)
add_value_labels(bars2)

# Adjust the subplot layout to give more room for labels
fig.tight_layout()

# Display the plot
plt.show()
```
